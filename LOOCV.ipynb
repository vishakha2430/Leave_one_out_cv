# Model building

```python
y=df['target']
```

```python
del df['target']
```

```python
x=df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 101)
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
```

# *Decision* Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

dt_classifier = DecisionTreeClassifier(random_state=42)



y_pred_train = dt_classifier.predict(x_train)
y_pred_test = dt_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print('Training Accuracy:', accuracy_train)
print('Test Accuracy:', accuracy_test)


if accuracy_train > accuracy_test:
    print('Warning: Model might be overfitting.')
else:
    print('Model performance is consistent between training and test data.')
dt_classifier.fit(x_train, y_train)

y_pred = dt_classifier.predict(x_test)


accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score
import numpy as np

dt_classifier = DecisionTreeClassifier(random_state=42)


loo = LeaveOneOut()


y_true = []
y_pred = []


for train_index, test_index in loo.split(x):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]


    print("Train Index:", train_index, "Test Index:", test_index)


    dt_classifier.fit(x_train, y_train)


    yhat = dt_classifier.predict(x_test)


    y_true.append(y_test.iloc[0])
    y_pred.append(yhat[0])


    accuracy = accuracy_score(y_true, y_pred)

    print('Accuracy for split {}: {:.4f}'.format(len(y_true), accuracy))

final_accuracy = accuracy_score(y_true, y_pred)
print('Final Accuracy:', final_accuracy)
```
